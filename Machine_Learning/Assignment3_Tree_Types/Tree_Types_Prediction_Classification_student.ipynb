## 3. Feature Engineering

### My Plan of Feature Extraction
- First, I decided to produce&transform a new column with ``Horizontal_Distance_To_Hydrology`` and ``Vertical_Distance_To_Hydrology`` columns. New column will contain the values of **Hypotenuse** of ``horizantal`` and ``vertical`` distances.

- As second, we can produce&transform an additional column which contains **average** of Horizantal Distances to Hydrology and Roadways.

- Third, I decided to transform a new column which contains **average** of `Elevation` and ``Vertical_Distance_To_Hydrology`` columns. So that, there is no need to have ``Horizontal_Distance_To_Hydrology`` and ``Vertical_Distance_To_Hydrology`` columns, because I have new columns which represent more value than them. I decide to drop these columns.

- Note that, after seeing the result of the models, there may be a possibility of making minor changes to the features in the modeling phase.

```python
df["Vertical_Distance_To_Hydrology"]
```

```python
df.Elevation
```

```python
df["Hypotenuse_distance_to_hydrology"] = ( (df["Horizontal_Distance_To_Hydrology"])**2 + (df["Vertical_Distance_To_Hydrology"])**2 )//0.5
```

```python
df.head()
```

```python
df["average_of_Horiz_Dist_Hydrology_Roadways"] = (df["Horizontal_Distance_To_Hydrology"] + df["Horizontal_Distance_To_Roadways"])/2
```

```python
df.head()
```

```python
df.drop(["Horizontal_Distance_To_Hydrology"], axis=1, inplace=True)
df.drop(["Vertical_Distance_To_Hydrology"], axis=1, inplace=True)
```

### Implement Support Vector Machine
- Import the modul 
- Fit the model

- Visualize and evaluate the result (use yellowbrick module)

*Note: You probably won't get a successful result. You may need to make some changes to the model or data. This may be a topic worth investigating, you decide.*

```python
def eval_metric(model, X_train, y_train, X_test, y_test):
    y_train_pred = model.predict(X_train)
    y_pred = model.predict(X_test)
    
    print("Test_Set")
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    print()
    print("Train_Set")
    print(confusion_matrix(y_train, y_train_pred))
    print(classification_report(y_train, y_train_pred))
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
svc = LinearSVC(random_state=42)
```

```python
svc.fit(X_train_scaled,y_train)
```

```python
y_pred = svc.predict(X_test)
```

```python
eval_metric(svc, X_train_scaled, y_train, X_test_scaled, y_test)
```

```python
y_pred_test = svc.predict(X_test_scaled)
y_pred_train = svc.predict(X_train_scaled)
```

```python
svc_accuracy_test = accuracy_score(y_test, y_pred_test)
svc_accuracy_train = accuracy_score(y_train, y_pred_train)

svc_f1_test = f1_score(y_test, y_pred_test, average='macro')
svc_f1_train = f1_score(y_train, y_pred_train, average='macro')

svc_accuracy_test, svc_accuracy_train, svc_f1_test, svc_f1_train
```
